4. If the electron in the ground state of hydrogen atom were to go around the hydrogen nucleus
following Bohr model, approximately how many rounds will it make in the time estimated in
the question above (c).<br>
**Answer:**<br><br>
Unielectronic species follow Bohr's Model. According to the Bohr's Model:<br><br>
**Velocity of an electron** = $\boxed{2.18 \cdot 10^{8} \frac{Z}{n} \frac{cm}{s}}$ <br><br>
**Radius of an orbit** = $\boxed{0.529 \cdot \frac{n^{2}}{Z} \AA}$<br><br>
**Derivation:**

In Bohr's model of the hydrogen atom, the electron orbits the nucleus in a circular orbit. The centripetal force required for this motion is provided by the electrostatic attraction between the electron and the nucleus. The centripetal force is given by:
$$
F_{\text{centripetal}} = \frac{mv^2}{r}
$$ <br>

where $m$ is the mass of the electron, $v$ is its velocity, and $r$ is the radius of the orbit.

The electrostatic attraction between the electron and the nucleus is given by Coulomb's law:

$$
F_{\text{electrostatic}} = \frac{1}{4\pi\varepsilon_0} \cdot \frac{e^2}{r^2}
$$

where $e$ is the elementary charge and $\varepsilon_0$ is the vacuum permittivity.

For stable equilibrium, the centripetal force and the electrostatic force must be equal:

$$
\frac{mv^2}{r} = \frac{1}{4\pi\varepsilon_0} \cdot \frac{e^2}{r^2}
$$

Solving for $v$:

$$
v = \sqrt{\frac{e^2}{4\pi\varepsilon_0 m r}}
$$

Bohr's model also quantizes angular momentum:

$$
L = mvr = n\hbar
$$

where $n$ is an integer and $\hbar$ is the reduced Planck constant.

Substituting $v$ from the earlier equation:

$$
mvr = n\hbar
$$

Solving for $v$:

$$
v = \frac{n\hbar}{mr}
$$

Substituting this back into the previous expression for $v$:

$$
v = \sqrt{\frac{e^2}{4\pi\varepsilon_0 m r}} = \frac{n\hbar}{mr}
$$

Solving for $r$:

$$
r = \frac{n^2 \hbar^2}{4\pi\varepsilon_0 e^2}
$$

#### Solution:

This is the expression for the radius of the \(n\)-th orbit in Bohr's model.<br>
For the Hydrogen Atom's Ground State, $n=1$ and $Z=1$, plugging in these values in the above obtained expressions for $v$ and $r$, the obtained values are:<br>
$$
v = 2.18 \cdot 10^{8} \frac{cm}{s}
$$
$$
r =  0.529 \cdot 10^{-10} m
$$
Distance covered in one round of the ground state = $ 2 \cdot \pi \cdot r $<br>
Distance that electron will cover in $10^{6}$ seconds with the above velocity $v$ = $ 10^{6} \cdot 2.18 \cdot 10^{8} \, \text{cm} = 2.18 \cdot 10^{12} \, \text{m} $<br>
Number of rounds = $\frac{2.18 \, \cdot \, 10^{12}}{2 \, \cdot \, \pi \,  \cdot \, 0.529 \, \cdot 10^{-10}}$<br>
$$
\boxed{\text{Final Answer} = 6560 \, \cdot \, 10^{18} \, \text{Rounds}}
$$

### Q2: Draw phase plane trajectories with three diferent values of total energies for a dynamical (mechanical) system


1. System Exhibiting SHM<br>
**Answer**:<br><br>
We know that, $\boxed{\text{T.E. = K.E. + P.E.}}$ <br>
where: $\text{T.E. = Total Energy of the System}$, $\text{K.E. = Kinetic Energy}$, $\text{P.E. = Potential Energy}$<br><br>
Consider a system with Potential Energy function given as $U(x) = \frac{k \cdot x^{2}}{2}$, and Total Energy denoted by $\text{T}$.<br>
Let us plot the Potential Energy Diagram for such a system with $k = 4$ for three different values of Total Energies, $ \text{T} = 2, 4, 6$.<br>


```python
import numpy as np
import matplotlib.pyplot as plt

k = 4
total_energies = [2, 4, 6]

def potential_energy(x):
    return 0.5 * k * x**2

x_values = np.linspace(-2, 2, 400)
potential_energy_values = potential_energy(x_values)

plt.figure(figsize=(7, 6))
plt.plot(x_values, potential_energy_values, label='Potential Energy Curve')

for energy in total_energies:
    plt.axhline(y=energy, color='r', linestyle='--', label=f'Total Energy = {energy}')

plt.title('Potential Energy vs. $x$')
plt.xlabel('Distance $x$')
plt.ylabel('Potential Energy')
plt.legend()
plt.grid()
plt.show()
```

To plot the phase-plane diagram of the system for the aforementioned total energies, we need to find the expression of momentum $p$ as a function of $x$.<br>
$$
\frac{m \cdot v^{2}}{2} = T - \frac{k \cdot x^{2}}{2}
$$
$$
v = \pm \sqrt{\frac{2}{m} \cdot (T - \frac{k \cdot x^{2}}{2})}
$$
$$
\boxed{p = \pm \sqrt{2 \cdot m \cdot (T - \frac{k \cdot x^{2}}{2})}}
$$
Plotting the phase-plane diagram for $ m = 1$, $k = 4$ and $T = 2, 4, 6$:


```python
import numpy as np
import matplotlib.pyplot as plt

m = 1
k = 4
total_energies = [2, 4, 6]

def calculate_a(total_energy):
    a = np.sqrt(2*total_energy/k)
    return a

def calculate_b(total_energy):
    b = np.sqrt(2*total_energy * m)
    return b

theta = np.linspace(0, 2*(np.pi), 500);
plt.figure(figsize=(10, 6))

for energy in total_energies:
    x = calculate_a(energy) * np.cos(theta)
    y = calculate_b(energy) * np.sin(theta)
    plt.plot(x, y, label=f'Total Energy = {energy}')

plt.title('Phase-Plane Diagram')
plt.xlabel('Position (x)')
plt.ylabel('Momentum (p)')
plt.legend()
plt.grid()
plt.show()
```

2. System is subject to an anharmonic potential with the energy in one case higher than the dissociation
limit. <br>
**Answer**: An anharmonic potential is a potential energy function that is not quadratic or linear, meaning it's non-linear. We shall proceed with the following anharmonic potential energy function:
$$ U(x)=5\left(1-e^{-\left(x-6\right)}\right)^{2} $$ 
The following is the graph for the function: <br>

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return 5 * (1 - np.exp(- (x - 6)))**2

x_values = np.linspace(5.2, 12, 400)
y_values = func(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values)
plt.title('Plot of $U(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

From the Figure, it is evident that the dissociation limit for the above system is at $\boxed{\text{Energy} = 5 \, \text{units}}$.<br>
Let us now plot the phase-plane trajectory of a system subjected to this potential energy for three different values of total energies: $\text{T} = 4, 5, 6$:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.logspace(0, 1.15, 1000000)

total_energies = [4, 5, 6]
color = ['orange', 'blue', 'green']

def calc_y(T):
    y_positive = np.sqrt(2 * (T - 5 * (1 - np.exp(-(x - 6)))**2))
    return y_positive

i=0
for energy in total_energies: 
    y_positive = calc_y(energy)
    y_negative = -y_positive
    plt.scatter(x, y_positive, c=color[i], marker='o', label=f'Total Energy = {energy}', linewidths=0.1, s=10)
    plt.scatter(x, y_negative, c=color[i], marker='o', linewidths=0.1, s=10)
    i+=1

plt.xlabel('Distance $(x)$')
plt.ylabel('Momentum $(p)$')
plt.title('Phase-Plane Trajectory of an Anharmonic System')
plt.legend()
plt.grid(True)
plt.show()
```

Thus, we have obtained the phase-plane trajectory for three different values of total energy. One lower than the dissociation limit ($T < 5$), one equal to the dissociation limit ($T = 5$) and one higher than the dissociation limit ($T > 5$).

### Q3: The logistic map $x_{n+1} = \alpha x_{n} (1 − x_{n})$ is usually studied for the growth parameter, $\alpha$, in the range $0 ≤ \alpha ≤ 4$. Why? Write a small code for the logistic map and compare the results graphically for $\alpha = 0.8$ and $\alpha = 1.5$ for initial conditions $x_{0} = 0.1$ and $0.5$. Take another case with $\alpha = 3.2$ and $x_{0} ≳ 0$ and plot the results.

**Answer:** <br>
The logistic map is studied for the following range of the growth parameter $0 ≤ \alpha ≤ 4$ because within this range, the logistic map exhibits a variety of dynamical behaviors, including stable fixed points, periodic orbits, and chaotic behavior, depending on the value of α and the initial conditions.

When $\alpha$ is smaller than $1$, the population approaches an equilibrium point, showing damped oscillations. When $\alpha$ is between $1$ and $3$, the population converges to a periodic cycle. However, when $\alpha$ exceeds $3.57$, the map begins to show chaotic behavior.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(alpha, x, maximum_n):
    results = [x]
    for _ in range(maximum_n):
        x = alpha * x * (1 - x)
        results.append(x)
    return results

alpha_values = [0.8, 1.5]
initial_conditions = [0.1, 0.5]
maximum_n = 50

plt.figure(figsize=(10, 6))

for alpha in alpha_values:
    for x0 in initial_conditions:
        results = logistic_map(alpha, x0, maximum_n)
        plt.plot(results, label=f'alpha={alpha}, x0={x0}')

plt.title("Logistic Map")
plt.xlabel("Step Value (n)")
plt.ylabel("Value of x_n")
plt.legend()
plt.grid(True)
plt.show()
```

For the case of $\alpha=3.2$ and $x_{0}≳ 0$ :

```python
initial_value = 1e-10
results = logistic_map(3.2, initial_value, maximum_n)
plt.plot(results)

plt.title("Logistic Map")
plt.xlabel("Step Value (n)")
plt.ylabel("Value of x_n")
plt.grid(True)
plt.show()
```

### Q4:  Show that in the limit of large $N$, the binomial distribution of $n$ out of $N$ objects becomes a Gaussian distribution.

**Answer:** <br>
Let $F(n)=\binom{N}{n}= \frac{N!}{n! \cdot (N-n)!}$<br><br>
We start by maximizing $F(n)$. To do that, we shall take the natural logarithm on both sides of the equation. Since $ln(x)$ is a increasing function, maximizing $ln(F(n))$ is akin to maximizing $F(n)$ $\rightarrow$ <br><br>
$\ln\left(F\left(n\right)\right)=\ln\left(\frac{N!}{n!\cdot\left(N-n\right)!}\right)=\ln\left(N!\right)-\ln\left(n!\right)-\ln\left(\left(N-n\right)!\right)$<br><br>
Let us see what the **Stirling's Approximation** is: For a large enough $N$, $\ln\left(N!\right)=N\ln\left(N\right)-N$<br><br>

**The assumption:** We are looking for the value of $n$ in the vicinity of the most probable distribution such that it maximizes $F(n)$. For this/these values of $n$, we may assume that $n$ and $N-n$ are also a large number such that the Stirling's approximation can be applied on them.<br><br>
$\ln\left(F\left(n\right)\right)=N\ln\left(N\right)-N\ -\ \left(n\ln\left(n\right)-n\right)-\left(\left(N-n\right)\ln\left(N-n\right)-\left(N-n\right)\right)$<br>
$\ln\left(F\left(n\right)\right)=N\ln\left(N\right)-\ n\ln\left(n\right)\ -\ \left(N-n\right)\ln\left(N-n\right)$<br>

To find the value of $n$ that maximizes the value of $F(n)$, we shall differentiate the above equation w.r.t $n$ and equate it to $0$ to find the corresponding value of $n$.<br>

$\dfrac{dF(ln(n))}{dn} = 0\ -\ \frac{N-n}{N-n}\left(-1\right)-\ln\left(N-n\right)\left(-1\right)-\frac{n}{n}-\ln\left(n\right) = 0$<br>

Let us call the $n$ that maximizes $F(n)$ as $n^{*}$. Therefore we have $n^{*}=N-n^{*} \rightarrow n^{*}=\frac{N}{2}$

Since the value of $n^{*}$ is $\frac{N}{2}$, it is a large enough number and our above assumption turns out to be correct, we can now proceed with the rest of the proof. <br><br>
The Taylor series expansion of $f(x)$ centered at $x = a$ is given by:
$f(x) = f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f'''(a)}{3!}(x - a)^3 + \dotsb$<br><br>
Let us find the Taylor Series expansion for $f(n)=ln(F(n))$ centered around $n^{*}$ up to the second derivate. The further derivates are negligible and hence not considered for the purposes of this proof.<br><br>
$f\left(n\right)=f\left(n^{*}\right)+\left(n-n^{*}\right)f'\left(n^{*}\right)+\frac{\left(n-n^{*}\right)^{2}}{2!}f''\left(n^{*}\right)$<br>
At $n=n^{*}$, the first derivative of the function $f(n)=ln(F(n))$ is equal to zero (since it is an extremum point).<br>
$f''(n)=\dfrac{df'(n)}{dn}$<br>
We know from above calculations that, $f'(n)=ln(N-n)-ln(n)$, therefore $f''(n)=\frac{-1}{N-n}-\frac{1}{n}$
At $n=n^{*}=\frac{N}{2}$, $f''(n^{*})= \frac{-4}{N}$<br><br>
We now have, $f\left(n\right)=f\left(n^{*}\right)+\frac{\left(n-n^{*}\right)^{2}}{2}\left(-\frac{4}{N}\right)$. Taking antilog on both sides...<br>
$F\left(n\right)=F\left(n^{*}\right)\exp\left[-\frac{2}{N}\cdot\left(n-n^{*}\right)^{2}\right]$<br>
The above formula resembles the gaussian distribution which is written as: $f(x) = \frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}$.<br><br>

Thus, we can say that in the limit where $N$ is large, the binomial distribution becomes a gaussian distribution.
